```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', encoding='windows-1251', index_col='laptop_id')
df.head()
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu'].value_counts()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda row: row.split()[0])
df.head()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
df['Memory'].value_counts()
```

```python
# 256GB SSD +  1.0TB Hybrid 
def sum_gb(row):
    lst_mem = row.split()
    gb = 0
    for el in lst_mem:
        if 'GB' in el:
            gb += int(el[:-2])
        elif 'TB' in el: 
            gb += float(el[:-2]) * 1024
    return gb
    
df['Memory_Amount'] = df['Memory'].apply(sum_gb)
df
```

```python
df['Memory_Amount'].max()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
# 256GB SSD +  1.0TB Hybrid 
def memory_type(row):
    lst_mem = row.split()
    el_type = ''
    for el in lst_mem:
        if el.isalpha():
            el_type += el + ' '
    return el_type.strip()
    
df['Memory_Type'] = df['Memory'].apply(memory_type)
df.sample(20)
```

```python
# 256GB SSD +  1.0TB Hybrid 
def memory_type_2(row):
    lst_mem = row.split()
    for i in range(len(lst_mem)-1, -1, -1):
        if 'GB' in lst_mem[i] or 'TB' in lst_mem[i]:
            del lst_mem[i]
    return ' '.join(lst_mem)
    
df['Memory_Type_2'] = df['Memory'].apply(memory_type_2)
df.sample(20)
```

### 1.4 Удалите признаки Memory и ScreenResolution

```python
df
```

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace=True)
df
```

## Задание 4.
Составьте несколько сводных таблиц
### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('Company').agg({'Price_euros': 'mean'}).sort_values('Price_euros')
```

### 4.2 Найдите минимальную, среднюю и максимальную
стоимости ноутбуков в зависимости от производителя
процессора

```python
df.groupby('Cpu_Company').agg({'Price_euros': ['min', 'mean', 'max']})
```

### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
df.pivot_table(index='Cpu_Company', columns='Ram', aggfunc = 'count', values='Weight', fill_value=0)
```

### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
df.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'}).unstack('Memory_Amount').fillna(0).round(2)
```
