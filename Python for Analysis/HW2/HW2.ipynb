### 5.2 –í—ã–±–µ—Ä–∏—Ç–µ –¥–æ–º –∫–ª–∏–µ–Ω—Ç—É
–ö–ª–∏–µ–Ω—Ç —Ö–æ—á–µ—Ç –¥–æ–º –ª–∏–±–æ —Å –æ—á–µ–Ω—å –∫—Ä–∞—Å–∏–≤—ã–º –≤–∏–¥–æ–º –∏–∑ –æ–∫–Ω–∞, –ª–∏–±–æ —Å
